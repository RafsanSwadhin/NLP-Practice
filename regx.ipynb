1. Extract all twitter handles from following text. Twitter handle is the text that appears after https://twitter.com/ and is a single word. Also it contains only alpha numeric characters i.e. A-Z a-z , o to 9 and underscore _

````python
import re
text = '''
Follow our leader Elon musk on twitter here: https://twitter.com/elonmusk, more information 
on Tesla's products can be found at https://www.tesla.com/. Also here are leading influencers 
for tesla related news,
https://twitter.com/teslarati
https://twitter.com/dummy_tesla
https://twitter.com/dummy_2_tesla
'''
pattern = 'https://twitter\.com/([a-zA-Z0-9_]+)'

re.findall(pattern, text)
````

explanation : 

Great question!

In the pattern:

```python
pattern = 'https://twitter\.com/([a-zA-Z0-9_]+)'
```

the backslash (`\`) before the dot (`.`) in `twitter\.com` is **used to escape the dot**.

---

### Why?

In **regular expressions**, the dot `.` is a **special character** that matches **any single character** (except a newline).
So:

* `twitter.com` (without backslash) would match:

  * `twitterXcom`, `twitter-com`, `twitter1com`, etc.
* `twitter\.com` (with backslash) means:

  * match a **literal dot** (`.`)

---

### Summary:

* ‚úÖ `twitter\.com` = matches the exact string `twitter.com`
* ‚ùå `twitter.com` (unescaped) = matches things like `twitterXcom`, `twitteracom`, etc.

So the `\.` ensures precision. Without it, your regex might match incorrect or unexpected URLs.

2. Extract Concentration Risk Types. It will be a text that appears after "Concentration Risk:", In below example, your regex should extract these two strings

(1) Credit Risk

(2) Supply Rish

````python
text = '''
Concentration of Risk: Credit Risk
Financial instruments that potentially subject us to a concentration of credit risk consist of cash, cash equivalents, marketable securities,
restricted cash, accounts receivable, convertible note hedges, and interest rate swaps. Our cash balances are primarily invested in money market funds
or on deposit at high credit quality financial institutions in the U.S. These deposits are typically in excess of insured limits. As of September 30, 2021
and December 31, 2020, no entity represented 10% or more of our total accounts receivable balance. The risk of concentration for our convertible note
hedges and interest rate swaps is mitigated by transacting with several highly-rated multinational banks.
Concentration of Risk: Supply Risk
We are dependent on our suppliers, including single source suppliers, and the inability of these suppliers to deliver necessary components of our
products in a timely manner at prices, quality levels and volumes acceptable to us, or our inability to efficiently manage these components from these
suppliers, could have a material adverse effect on our business, prospects, financial condition and operating results.
'''
pattern = 'Concentration of Risk: ([^\n]*)'

re.findall(pattern, text)
````

explanation:


```python
pattern = 'Concentration of Risk: ([^\n]*)'
```

### üîç Explanation

#### 1. **Literal part: `'Concentration of Risk: '`**

* This matches **exactly** the phrase:

  ```
  Concentration of Risk:
  ```
* It is **case-sensitive** unless you specify otherwise in the regex options (e.g., `re.IGNORECASE`).

#### 2. **Capturing Group: `([^\n]*)`**

This part is very important:

* `(` and `)` ‚Üí This is a **capturing group**, meaning this part of the match will be **extracted**.
* `[^...]` ‚Üí This is a **negated character class**, meaning ‚Äúmatch anything **except** the characters listed‚Äù.
* `\n` ‚Üí This is a **newline character**.
* `*` ‚Üí Match **zero or more** of the previous character class.

So together:

### üìå `([^\n]*)` means:

> "Capture everything **after** 'Concentration of Risk: ' until the **end of the line** (i.e., before the next `\n`)."

---

### ‚úÖ Example:

```python
import re

text = """
Company Overview
Concentration of Risk: High due to single supplier.
Financials
"""

pattern = r'Concentration of Risk: ([^\n]*)'

match = re.findall(pattern, text)
print(match)
```

### Output:

```python
['High due to single supplier.']
```


3. Companies in europe reports their financial numbers of semi annual basis and you can have a document like this. To exatract quarterly and semin annual period you can use a regex as shown below

Hint: you need to use (?:) here to match everything enclosed

````python
text = '''
Tesla's gross cost of operating lease vehicles in FY2021 Q1 was $4.85 billion.
BMW's gross cost of operating vehicles in FY2021 S1 was $8 billion.
'''

pattern = 'FY(\d{4} (?:Q[1-4]|S[1-2]))'
matches = re.findall(pattern, text)
matches
````

---

### üîç Step-by-Step Explanation

#### üìÑ `text = ''' ... '''`

You have a multiline string containing financial data for **Tesla** and **BMW**, with fiscal year and quarter/semester indicators like `FY2021 Q1` and `FY2021 S1`.

---

#### üß† Regular Expression:

```python
pattern = 'FY(\d{4} (?:Q[1-4]|S[1-2]))'
```

Let‚Äôs break this down piece by piece:

##### üß© `'FY'`

* Matches the **literal string** `'FY'`, which stands for **Fiscal Year**.

##### üß© `(\d{4} (?:Q[1-4]|S[1-2]))`

This is the **capturing group** that we want to extract.

###### üî¢ `\d{4}`

* `\d` means **any digit** (`0‚Äì9`).
* `{4}` means **exactly 4 digits**.
* So this matches the **year**, e.g., `2021`.

###### üß© `' '` (space)

* A space between the year and quarter/semester.

###### üß© `(?:Q[1-4]|S[1-2])`

* This is a **non-capturing group** (`(?:...)`) used for grouping without capturing.
* It matches either:

  * `Q[1-4]` ‚Üí A quarter (Q1, Q2, Q3, Q4)
  * `S[1-2]` ‚Üí A semester (S1 or S2)

So `(?:Q[1-4]|S[1-2])` matches any one of:

* `Q1`, `Q2`, `Q3`, `Q4`
* `S1`, `S2`

---

### ‚úÖ What does the pattern match?

The entire regex:

```regex
FY(\d{4} (?:Q[1-4]|S[1-2]))
```

Matches strings like:

* `FY2021 Q1` ‚Üí captures `2021 Q1`
* `FY2021 S1` ‚Üí captures `2021 S1`

The `FY` part is not captured (it's outside the parentheses), but the rest (`2021 Q1`, `2021 S1`) **is captured**.

---

### üì¶ Output of `re.findall(...)`

Given your `text`, the matches will be:

```python
['2021 Q1', '2021 S1']
```
